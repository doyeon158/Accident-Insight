```python
# 음식점 개수
df_nonalcohol_count = df_nonalcohol.groupby(['시군구명','행정동명']).count().reset_index()
df_nonalcohol_count = df_nonalcohol_count[['시군구명','행정동명','상호명']]
df_nonalcohol_count = df_nonalcohol_count.rename(columns={'상호명':'음식점 개수'})
df_nonalcohol_count.head()
```

```python
# 음식점 행정동 코드 결합
def find_code(row, df_nonalcohol):
  match = df_nonalcohol[(df_nonalcohol['행정동명'] == row['행정동명']) &
                          (df_nonalcohol['시군구명'] == row['시군구명'])]
  if not match.empty :
      return match['행정동코드'].values[0]
  else :
      return None

df_nonalcohol_count['행정동코드'] = df_nonalcohol_count.apply(lambda row: find_code(row, df_nonalcohol), axis=1)
df_nonalcohol_count.head()
```

```python
df_nonalcohol_count['행정동코드'] = df_nonalcohol_count['행정동코드'].astype(str) + '00'
df_nonalcohol_count.head()
```
