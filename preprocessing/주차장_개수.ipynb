# 주차장 데이터

```python
df_parking_lot = pd.read_csv('/content/drive/MyDrive/[통계청 공모전] 음주운전/1 데이터/소현_유동인구 상권 및 시설 위치/주차장 위치/주차장(동별)(2016년+이후)_20240619232256.csv')
df_parking_lot.head()
```

```python
# 구, 동, 가장 최근 주차장 총 개수 데이터만 남기기
df_parking_lot = df_parking_lot.iloc[:,1:4]
df_parking_lot.columns = ['구','동', '주차장 개수']

# 값만 남기기
df_parking_lot.drop(index = range(0,4), axis = 0, inplace = True)
df_parking_lot.head()
```

```python
# 소계 빼기
df_parking_lot = df_parking_lot[df_parking_lot['동'].str.endswith('동')]
df_parking_lot.head()
```

# 행정동 코드 결합

```python
# 행정동 코드
df_code = pd.read_csv('/content/drive/MyDrive/[통계청 공모전] 음주운전/1 데이터/소현_유동인구 상권 및 시설 위치/행정동및법정동코드.csv')
df_code.head()
```

```python
# 행정동 코드 결합
def find_code(row, df_code) :
  match = df_code[df_code['읍면동명']==row['동']]
  if match.empty :
    return None
  else :
    return match['행정동코드'].values[0]

df_parking_lot['행정동 코드'] = df_parking_lot.apply(lambda row : find_code(row, df_code), axis = 1)
df_parking_lot.reset_index(drop = True, inplace = True)
df_parking_lot.head()
```

## 결측치 처리

```python
# 결측치
import re
def add_prefix_to_dong(dong):
    if dong and isinstance(dong, str):  # dong 값이 문자열인 경우
        return re.sub(r'(\D+)(\d+)', r'\1제\2', dong)
    return dong
nan_rows = df_parking_lot[df_parking_lot.isnull().any(axis=1)]
df_parking_lot.loc[nan_rows.index, '동'] = nan_rows['동'].apply(add_prefix_to_dong)
df_parking_lot[df_parking_lot.isnull().any(axis=1)]
```

```python
# 결측치 행정동 코드 다시 결합
def find_code(row, df_code) :
  match = df_code[df_code['읍면동명']==row['동']]
  if match.empty :
    return None
  else :
    return match['행정동코드'].values[0]

df_parking_lot['행정동 코드'] = df_parking_lot.apply(lambda row : find_code(row, df_code), axis = 1)
df_parking_lot[df_parking_lot.isnull().any(axis=1)]
```
