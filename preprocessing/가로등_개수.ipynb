# 가로등 위치 데이터

```python
df_light = pd.read_csv('/content/drive/MyDrive/[통계청 공모전] 음주운전/1 데이터/소현_유동인구 상권 및 시설 위치/가로등 위치/서울특별시_가로등 위치 정보_20221108.csv', encoding = 'euc-kr')
df_light.head()
```

```python
# 행정동 경계 데이터
gdf_dong = gpd.read_file('/content/drive/MyDrive/[통계청 공모전] 음주운전/1 데이터/소현_유동인구 상권 및 시설 위치/행정동경계.geojson')
```

```python
df_light['geometry'] = df_light.apply(lambda row: Point([row['경도'], row['위도']]), axis=1)
gdf_points = gpd.GeoDataFrame(df_light, geometry='geometry')
```

```python
gdf_points.crs = {'init': 'epsg:4326'}
gdf_dong = gdf_dong.to_crs(epsg=4326)
```

```python
gdf_result = gpd.sjoin(gdf_points, gdf_dong, how='left', op = 'within')
gdf_result.head()
```

```python
# 기존 데이터와 결합
df_light['행정동명'] = gdf_result['temp']
df_light['행정동 코드'] = gdf_result['adm_cd2']
df_light.head()
```

```python
df_light[df_light.isnull().any(axis=1)]
```

```python
# 결측치 drop
df_light.dropna(inplace=True)
df_light.reset_index(drop=True, inplace=True)
df_light.head()
```

```python
# 행정동별 가로등 개수 count
df_light_count =df_light.groupby('행정동명').count()
df_light_count.reset_index(inplace=True)
df_light_count = df_light_count[['행정동명', '행정동 코드']]
df_light_count.columns = ['행정동명', '가로등 개수']
df_light_count.head()
```

```python
# 행정동 코드 결합
def find_code(row, df_light):
    match = df_light[df_light['행정동명'] == row['행정동명']]
    if not match.empty:
        return match['행정동 코드'].values[0]
    else:
        return None

df_light_count['행정동 코드'] = df_light_count.apply(find_code, axis=1, args=(df_light,))

# 전처리 최종
df_light_count.to_csv('/content/drive/MyDrive/[통계청 공모전] 음주운전/1 데이터/소현_유동인구 상권 및 시설 위치/가로등 위치/행정동별 가로등 개수.csv')
df_light_count.head()
```
