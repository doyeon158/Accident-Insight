```python
df_pub_num = pd.read_csv('/content/drive/MyDrive/[통계청 공모전] 음주운전/1 데이터/소현_유동인구 상권 및 시설 위치/유동인구/서울시 행정동별 대중교통 총 승차 승객수 정보.csv',
                         encoding='euc-kr')
df_pub_num = df_pub_num.groupby('행정동_ID').mean()
df_pub_num.reset_index(inplace=True)
df_pub_num = df_pub_num.drop(['기준_날짜'], axis = 1)
df_pub_num.head()
```

```python
df_dong = pd.read_csv('/content/drive/MyDrive/[통계청 공모전] 음주운전/1 데이터/소현_유동인구 상권 및 시설 위치/유동인구/서울시 읍면동마스터 정보.csv', encoding = 'euc-kr')
df_dong.head()
```

```python
def find_dong(row, df_dong):
  match = df_dong[df_dong['행정동_ID'] == row['행정동_ID']]
  if not match.empty:
    return match['행정동_명칭'].values[0]
  else:
    return None

df_pub_num['행정동_명칭'] = df_pub_num.apply(find_dong, args=(df_dong,), axis=1)
df_pub_num.head()
```
