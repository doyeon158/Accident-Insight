```python
import pandas as pd
import matplotlib.pyplot as plt

# 예제 데이터 생성
data = pd.DataFrame({
    '자치구': ['강남구', '강서구', '서초구', '송파구', '은평구'],
    '예산': [500000, 400000, 300000, 200000, 100000]
})

# 자치구를 인덱스로 설정
data = data.set_index('자치구')

# 파이 차트 생성
plt.figure(figsize=(8, 6))
plt.pie(data['예산'], labels=data.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])

# 차트 제목 설정
plt.title('자치구별 예산 배분 비율')

# 파이 차트 보여주기
plt.axis('equal')  # 파이 차트를 원형으로 유지
plt.show()
```
